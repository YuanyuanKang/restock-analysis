```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create mock inventory dataset
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Current_Stock': [18, 25, 12, 30, 19],
    'Safety_Stock': [20, 20, 15, 25, 20]
}

df = pd.DataFrame(data)

# 2. Check whether restocking is needed
df['Restock_Needed'] = df['Current_Stock'] < df['Safety_Stock']

# 3. Print the result
print("üßæ Inventory Restocking Check Result:")
print(df)

# 4. Visualize the restocking needs
plt.figure(figsize=(8, 5))
df['Restock_Needed'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.title('Number of Products That Need Restocking')
plt.ylabel('Product Count')
plt.xlabel('Restock Status')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
